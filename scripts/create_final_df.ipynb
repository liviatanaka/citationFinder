```python
import pandas as pd
```

```python
arvix = pd.read_csv("../cleaned_data/arxiv.csv")
springer = pd.read_csv("../cleaned_data/publications_data.csv")
```

```python
arvix.head()
```

```python
arvix['source'] = 'arXiv'
```

```python
arvix
```

```python
springer['source'] = "Springer"
```

```python
arxiv = arvix.filter(["title", "abstract", "source"], axis=1)
springer = springer.filter(["title", "abstract", "source"], axis=1)
```

```python
merged = pd.concat([arxiv, springer])
```

```python
merged.shape
```

```python
merged.head()
```

```python
merged = merged[(merged["title"].notna()) & (merged["abstract"].notna())]
```

```python
merged.shape
```

```python
merged.to_csv("merged_publications.csv", index=False)
```
